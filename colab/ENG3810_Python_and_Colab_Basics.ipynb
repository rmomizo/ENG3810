```
#create a function that converts a string to lowercase and tokenizes it
def my_tokenizer(arg):
  """Function takes a string as arg and returns lower case word tokens""" #this is the docstring
  if type(arg) == str:
    return arg.lower().split()
  else:
    return print("invalid data type")
```

```
def test_tokenize(somestring):
  l = somestring.lower()
  s = l.split()
  tokens = []
  for word in s:
    if word not in stops:
      tokens.append(word)
    else:
      pass  
  final = []
  for tok in tokens:
    if len(tok) > 2:
      final.append(tok)
    else:
      pass

  return final
```

```
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.cluster import KMeans

#vectorizer word counts per document
cv = CountVectorizer()
tfidf = TfidfVectorizer()


cv.fit(tokens) #fit the vectorizer to the vocabulary
tfidf.fit(tokens)

cvecs = cv.transform(tokens) #transform original documents into count vectors

tfidf_vecs = tfidf.transform(tokens)
```

```
import matplotlib.pyplot as plt
import numpy as np
def get_most_informative_features(vectorizer, clf, feature_count=10):
    """Prints features with the highest coefficient values, per class"""
    class_labels = clf.classes_
    feature_names = vectorizer.get_feature_names()
    for i, class_label in enumerate(class_labels):
        top = np.argsort(clf.coef_[i])[1 * feature_count:]
        top_scores = np.argsort(clf.coef_[i][1 * feature_count:])
        print("%s: %s" % (top_scores,
              ", ".join(feature_names[j] for j in top)))
```
